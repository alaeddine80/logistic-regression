# 1-Preprocessing part


```python
#Importing the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics


```

```python
A=pd.read_csv(r"C:\\Users\LENOVO\Downloads\titanic.csv")
A
```

```python
#replacing missing value
A['Age'].fillna(A['Age'].mean(),inplace=True)
A
```

```python
#label enconding for 'Sex' feature
S=A['Sex']
s={'Sex':{'male':1,'female':0}}
A.replace(s,inplace=True)
A
```

```python
#one hot encoding for 'Embarked feature'
Z=pd.get_dummies(A['Embarked'])
A=A.drop('Embarked',axis=1)
A=A.join(Z)
A
```

```python
#filtering columns
A1=A.filter(['Name','Survived','Pclass','Sex','Age','SibSp','Parch','Fare','C','Q','S'])
A1
```

```python
Title=[]
for i in A1['Name'] :
    pos1 = i.find(',') 
    pos2 = i.find('.')
    sousChaine = i[pos1+2:pos2]
    Title.append(sousChaine)
    
Title
```

```python
A1['Title']=Title
A1
```

```python
A2=A1.filter(['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','C','Q','S','Title'])
A2
```

```python
#features extraction
x = A1[['Pclass','Sex','Age','SibSp','Parch','Fare','C','Q','S']]
y = A1['Survived']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)  #splitting data with test size of 25%

```

```python
logreg = LogisticRegression()   #build our logistic model
logreg.fit(x_train, y_train)  #fitting training data
y_pred  = logreg.predict(x_test)    #testing model’s performance
y_pred
```

```python

print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, y_pred)
print(cf_matrix)
```

```python
sns.heatmap(cf_matrix, annot=True)
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score 

logistic_fpr, logistic_tpr, threshold = roc_curve(y_test, y_pred)
auc_logistic = auc(logistic_fpr,logistic_tpr)

plt.figure(figsize=(10,5), dpi=60)
plt.plot(logistic_fpr, logistic_tpr, marker='.', label='Logistic (auc = %0.3f)' % auc_logistic)
plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')
plt.legend()
plt.show()
```

```python
auc = np.round(roc_auc_score(y_test,y_pred), 3) 
auc
```
